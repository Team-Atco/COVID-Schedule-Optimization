Here we define a loop to create a random schedule for one individual. Later we will create a loop for all the individuals.


```python


def create_day_schedule(work_hours=[7, 21], shift_length=[3, 5]):
    #Chooses a random start shift length for a day, returns them in a list
    np.random.seed()
    start=random.randint(work_hours[0], work_hours[1]-4)
    shift=random.randint(shift_length[0], min(shift_length[1],work_hours[1]-start ))
    return [start, shift]


def create_worker_schedule():
    #Returns a list [MonStart, MonShiftLength,...,]
    schedule=[]
    for day in ['Mon', 'Tue', 'Wed', 'Thur', 'Fri']:
        x=create_day_schedule()
        schedule.append(x[0]) 
        schedule.append(x[1])
    return schedule

def create_workforce_schedule(n=200, alpha=35):
    cols=['Monday Start Time', 'Monday Shift Length', 'Tuesday Start Time', 'Tuesday Shift Length', 'Wednesday Start Time', 'Wednesday Shift Length', 'Thursday Start Time', 'Thursday Shift Length', 'Friday Start Time', 'Friday Shift Length']
    df= pd.DataFrame({cols[0]: range(0, n), cols[1]: range(0, n), cols[2]: range(0, n), cols[3]: range(0, n), cols[4]: range(0, n), cols[5]: range(0, n), cols[6]: range(0, n), cols[7]: range(0, n), cols[8]: range(0, n), cols[9]: range(0, n)})
    #number_of_zeros=int(n-35/100*n)
    for employee in range(0, n):
        schedule=create_worker_schedule()
        for i in range(0, 10):
            df[cols[i]][employee]=schedule[i]
    
    return df
 
    
    
def count_employee_hours(df):
    x=[]
    for i in range(0, len(df['Monday Start Time'])):
        x.append(df['Monday Shift Length'][i]+df['Tuesday Shift Length'][i]+df['Wednesday Shift Length'][i]+df['Thursday Shift Length'][i]+df['Friday Shift Length'][i])
    print(x)
    
        
```

```python
def generate_population(P=40000, WorkForceSize=200):
    x=[]
    for i in range(0, P):
        y=create_workforce_schedule(n=WorkForceSize)
        x.append(y)
    return x
```

```python
def mate(A,B, mutationprob=0.15):
    fitA=objective_function(A)
    fitB=objective_function(B)
    prob=fitB/(fitA+fitB)
    C=A.copy()
    for employee in range(0, count_employees(A)):
        value=random.random()
        value2=random.random()
        if value>prob:
            C.iloc[employee]=B.copy().iloc[employee]
        if value2<mutationprob:
            newemployee=create_worker_schedule()
            C.iloc[employee]=newemployee
        

    
    return C
            
        
```

```python
def compare_objective(A,B,C):
    #Compares A B and C and returns 'A', 'B', or 'C' representing whichever letter gets thrown out
    fitA=objective_function(A)
    fitB=objective_function(B)
    fitC=objective_function(C)

    if fitC>fitA and fitC>fitB:
        return 'C'
    elif fitB>fitC:
        return 'B'
    else:
        return 'A'
    
    
```

```python
def selection(population):
    #Takes in a population, mates its members, and returns a new population of the same size based on fitness
    new_pop=[]
    P=len(population)
    for i in range(0, int(P/2)):
        Aindex=random.randint(0, P-1)
        Bindex=random.randint(0, P-1)
        A=population[Aindex]
        B=population[Bindex]
        C=mate(A,B)
        dead=compare_objective(A,B,C)
        if dead=='A':
            new_pop.append(B)
            new_pop.append(C)
        elif dead=='B':
            new_pop.append(A)
            new_pop.append(C)
        else:
            new_pop.append(A)
            new_pop.append(B)
    return(new_pop)
        
        
```

```python
def find_best(population):
    #Compares all members of a population and returns the best member
    P=len(population)
    best=population[0]
    fitness=objective_function(best)
    for member in population:
        if objective_function(member)< fitness:
            best=member
            fitness=objective_function(best)
    return best
```

```python
def genetic_algorithm(popsize=6, gens=3, NumEmployee=200):
    #Creates a population of size P, lets it survive over gens generations
    #Returns the max fitness in the first generation, the max fitness in the last generation, and the best member
    init_pop=generate_population(P=popsize, WorkForceSize=NumEmployee)
    init_best=find_best(init_pop)
    init_best_fitness=objective_function(init_best)
    population=init_pop
    for i in range(0, gens):
        population=selection(population)
    best=find_best(population)
    best_fitness=objective_function(best)
    return [init_best_fitness, best_fitness, best]
```

```python
p_list= [200]
g_list= [1000]
e_list= [20]
for p in p_list:
    for g in g_list:
        for e in e_list:            
            solution=genetic_algorithm(popsize=p, gens=g, NumEmployee=e)
            fileName='pop_'+str(p)+'_gen_'+str(g)+'_emp_'+str(e)+'.csv'
            df_result = solution[2]
            df_result = df_result.append({'First' : solution[0], 'Last' : solution[1]},   ignore_index = True) 
            df_result.to_csv(fileName)
```
